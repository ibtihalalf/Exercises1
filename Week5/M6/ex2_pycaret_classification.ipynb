 **First experiment**:

```python
import pandas as pd
from pycaret.classification import *

# Load data
penguins = sns.load_dataset("penguins")

# Keep only needed columns
df1 = penguins[["flipper_length_mm", "bill_length_mm", "species"]].dropna()

# Setup
clf1 = setup(
    data=df1,
    target="species",
    session_id=42,
    verbose=False
)

# Compare models
best_model_1 = compare_models()

# Show CV results table
results_1 = pull()
print(results_1)

# Analyze model
plot_model(best_model_1, plot="confusion_matrix")
plot_model(best_model_1, plot="class_report")
plot_model(best_model_1, plot="auc")  

# Predictions on holdout set
preds_1 = predict_model(best_model_1)
print(preds_1.head())

# Save model
save_model(best_model_1, "penguins_species_model")
```
